```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv('mental-heath-in-tech-2016_20161114.csv')

df.head(10)
```

```python
#146,169, 287, 1146(empty) are common null values

df.isnull().sum()
```

```python
df = df[df['How many employees does your company or organization have?'].notna()]
df.isnull().sum()
```

```python
print(list(df.isnull().sum()))
```

```python
df=df.dropna(axis=1,how='all')

print(list(df.isnull().sum()))
```

```python
df.isnull().sum()
```

```python
df = df[df['Have your previous employers provided mental health benefits?'].notna()]

print(list(df.isnull().sum()))
```
